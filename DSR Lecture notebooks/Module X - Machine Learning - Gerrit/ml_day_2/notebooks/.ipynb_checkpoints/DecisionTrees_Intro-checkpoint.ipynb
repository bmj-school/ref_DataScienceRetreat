## Decision Tree with scikit-learn 

Decision Trees can be founded in the `tree` package of scikit-learn.

```python
from sklearn import tree
```

We transform our labels into buckets.

```python
from sklearn.preprocessing import binarize

y_2c = binarize(y.reshape(-1, 1), np.median(y)).reshape(-1)
```

```python
print(y_2c)
print(type(y_2c))
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def do_cv_experiment(clf_to_score, X, y, k=5):
    scores = cross_val_score(clf, X, y_2c, scoring="roc_auc", cv=k, n_jobs=3)
    return np.mean(scores), np.std(scores)


mu, sigma = do_cv_experiment(clf, X, y_2c, 5)
print("%.4f (%.4f) AUC" % (mu, sigma))
```

```python
print(X.shape)
print(y_2c.shape)
```

```python
#print(clf.fit())
#clf.fit(X, y, sample_weight=None, check_input=True, X_idx_sorted=None)
fitted = clf.fit(X, y_2c)
print(fitted)
```

```python
for i in dir(fitted):
    print(i)
```

```python
df.head()
feature_names = list(df.columns.values)[0:-1]
feature_names
```

```python
import graphviza 
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=feature_names) 
graph = graphviz.Source(dot_data) 
graph.render("Big Tree") 
```
