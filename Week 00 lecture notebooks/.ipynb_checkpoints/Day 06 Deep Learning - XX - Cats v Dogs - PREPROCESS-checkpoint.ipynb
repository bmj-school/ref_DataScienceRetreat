```python
(train_images, train_labels), (test_images, test_labels) = ks.datasets.mnist.load_data()
print(type(train_images))
print(train_images.shape)
pd.DataFrame(train_images[0])
```

```python
# Reshape the data
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype("float32") / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype("float32") / 255
```

```python
train_labels = ks.utils.to_categorical(train_labels)
test_labels =  ks.utils.to_categorical(test_labels)
print(train_labels[0])
```

```python
model = ks.models.Sequential()
# Filters, kernal shape, activation, input shape
model.add(ks.layers.Conv2D(32, (3,3),activation = "relu", input_shape=(28,28,1)))
model.add(ks.layers.MaxPooling2D((2,2)))
model.add(ks.layers.Conv2D(64, (3,3),activation = "relu"))
model.add(ks.layers.MaxPooling2D((2,2)))
model.add(ks.layers.Conv2D(64, (3,3),activation = "relu"))
model.add(ks.layers.Flatten())
model.add(ks.layers.Dense(64, activation="relu"))
model.add(ks.layers.Dropout(0.5))
model.add(ks.layers.Dense(10,activation="softmax"))
#sum_str = model.summary()
config_dict = model.get_config()

pprint(config_dict)
print(type(config_dict))
print(len(config_dict))
```

```python
model.compile(   
    optimizer = ks.optimizers.RMSprop(lr=0.001),
    loss= ks.losses.binary_crossentropy,
    metrics= ["accuracy"],
)
```

```python
#model.summary()
```

```python
history = model.fit(
    train_images, train_labels,
    epochs = 5,
    batch_size = 64,
)
```

```python
pprint(history)
#print(history.__dict__)
history_dict = history.__dict__
pprint(history_dict)
#print(history_dict['epoch'])
#for k in history:
#    print(k)
```

```python
epochs = history_dict['epoch']
acc = history_dict['history']['acc']
loss = history_dict['history']['loss']
```
